```python
#load
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("sales_raw_500.xlsx")
print(df)
```

```python
#duplicates removal
df=df.drop_duplicates()
df #almost 20 rows are removed as duplicates 
```

```python
#fix date column bcoz there is a error in row 13
df['Order_Date']=pd.to_datetime(df['Order_Date'],errors='coerce')
```

```python
df.dropna(subset=['Order_Date'],inplace=True)
#save to df 
df
```

```python
df['Profit']=df['Sales']-df['Cost'] # create profit column
df['Year']=df['Order_Date'].dt.year # create year column
df['Month']=df['Order_Date'].dt.month # create month column
df
```

```python
df.to_csv("sales_cleaned.csv",index=False)
```
